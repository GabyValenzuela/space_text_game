<font color=black> **Welcome to the SpaceX-213 Game** </font>


<span><small>The purpose of this game is to take the user into a spaceship trip, which has the objective of reaching planet N20-XF809 that is supposed to be inhabitated by another intelligent life. The challenge of the game will be to make the user reach that planet or if that is not possible, to connect with this intelligent life and come back home with all its crew. During this trip the user will be given a couple of choices that will take them to a win or failure outcome.</small></span>

</span><small>This game is intended to be fun for the user and to explore different scenarios while playing it a second or third time.</small></span> 

</span><small>This game contains some truthful aspects such as the names of the planets, and some others based on the imagination of the creator.</small></span>

</span><small>Hope you enjoy the game!</small></span>


```python
"""
######################## WELCOME TO THE TEXT SPACESHIP GAME #############################

This is a text game in Python based on a spaceship mission. This game has 5 main stages:
        1 Mission Space
        2 Traveling Space
        3 Out of Via Lactea
        4 Planets
        5 Uncertainty
    As well as the Win (6) and Fail (7) scenarios

In the very first two stages, the user will keep advancing to other stages, and only as they 
start the 3rd stage (Out of Via Lactea) the game will start with the user making decisions about their journey.

I hope you have a good time playing this game! Comments and improvements are accepted.
"""

# Inserting packages used during this game
import time
import os, sys, time
import textwrap

# Defined function to slow text display letter by letter
def print_slow(s, speed, tail):
   for letter in s:
      print(letter, end="")
      sys.stdout.flush()
      time.sleep(speed)
   print(end=tail)

# Welcoming User
print("""
__        __   _                            _        
\ \      / /__| | ___ ___  _ __ ___   ___  | |_ ___  
 \ \ /\ / / _ \ |/ __/ _ \| '_ ` _ \ / _ \ | __/ _ \ 
  \ V  V /  __/ | (_| (_) | | | | | |  __/ | || (_) |
   \_/\_/ \___|_|\___\___/|_| |_| |_|\___|  \__\___/ 
 ____                      __  __    ____  _ _____   
/ ___| _ __   __ _  ___ ___\ \/ /   |___ \/ |___ /   
\___ \| '_ \ / _` |/ __/ _ \\  /_____ __) | | |_ \   
 ___) | |_) | (_| | (_|  __//  \_____/ __/| |___) |  
|____/| .__/ \__,_|\___\___/_/\_\   |_____|_|____/   
      |_|  _                 _                  _    
  __ _  __| |_   _____ _ __ | |_ _   _ _ __ ___| |   
 / _` |/ _` \ \ / / _ \ '_ \| __| | | | '__/ _ \ |   
| (_| | (_| |\ V /  __/ | | | |_| |_| | | |  __/_|   
 \__,_|\__,_| \_/ \___|_| |_|\__|\__,_|_|  \___(_)   
""")
time.sleep(1) 

welcome_speed = 0.03

print_slow(f"""
In this game you'll prove your ability as a commander of a spaceship.\n
Your mission if you accept it, is to reach planet N20-XF809 which is supposed to be
inhabited by another intelligent life and to make contact with them.\n
Either way, if you reach the planet or make contact with intelligent life your mission
will be completed.
""", welcome_speed, "\n\n")
time.sleep(1)
    
print_slow(f"""Are you ready to play?\n""", welcome_speed, "\n")
time.sleep(1)

#Making user start the game
print(f"""\n< Please type enter to start >""")
input(prompt = f""" """)
time.sleep(1)

#Defining speed for Mayday message and printing it
mayday_speed = 0.01
print_slow(f"""\n ...Mayday mayday we need assistance here!!!...""", mayday_speed, "\n\n")
time.sleep(2)

# Defining MISSION SPACE stage
def mission_space():
    #Defining global variable for user name
    global capitain_name

    #Message to user
    print_slow(f"""
    It's 2100 here on Earth, and you are the commander of Space X-213.\n
    Yes, the same type as that one launched by Tesla in 2020, but after 212 more launches\n\
    and many, many more improvements!
    
    Your mission as a commander is to arrive on a new planet called N20-XF809 which is\n\
    in Hoag's Object Galaxy.\n """, welcome_speed, "\n")

    #Giving the user a space for reading and interacting
    input(prompt = """\n Please type enter to continue \n""")
    
    #Message to user
    print_slow(f"""
    Your mission is to go there and investigate an intelligent life that sent a message
    two months ago and to come back safe and sound with all your crew.
    
    This trip won't be easy as a leader you'll have to make hard decisions that may
    or may not bring you back. """, welcome_speed, "\n")
    
    #Asking user's name
    print(f"""
    But before we start, please Captain, can you tell me your name: \n """)
    user_name = input(prompt = "") 
    capitain_name = user_name.capitalize()

    #Launching message
    launching_speed = 0.02

    print_slow(f"""
    \n 1 \n
    \n 2 \n
    \n 3...\n
    Launching Capitain {capitain_name}!!!""", launching_speed, "\n")
    
    #Giving the user a space for reading and interacting
    input(prompt = """ \n Please type enter to continue \n """)
    
    #Taking user to the following stage
    #traveling_space()
    print("END of mission stage")

#################################################################################
############################### TRAVELING SPACE STAGE ###########################
mission_space()
```

```python
  #################################################################################
############################### TRAVELING SPACE STAGE ###########################

### Defining stage... TRAVELING SPACE!!!
def traveling_space():
    ### Letting know the user that they are in the TRAVELING SPACE stage!
    print_slow(f"""
    You're finally in space and there have been a couple of hours since you left Earth...
    
    How does it make you feel?""","")
    
    ### Testing user excitement
    print_slow(f"""\n
    1. Excited! Never been better before!\n
    2. Fearful! What if something goes wrong?\n
    3. Not so sure...
    
    < Please type your answer > ""","")
    
    ###Conditional answer in regards to its level of excitement
    feeling = input(prompt = "")
    
    ###Adding rules so that the user can type a number or a word, even if a number was requested
    if feeling == "1" or feeling == "1." or feeling == "excited" or "excited" in feeling or "one" in feeling:
        print_slow(f"""
        GREAT! That is the adventure spirit!\n""","")
    elif feeling == "2" or feeling == "2." or feeling == "fearful" or "fearful" in feeling or "two" in feeling:
        print (f"""
        Well, you're right to be worried, lots of things can and will go wrong!\n""")
    ###Last condition that opens up so that the user that wasn't excited gets excited
    elif feeling == "3" or feeling == "3." or feeling == "not so sure" or "not sure" in feeling or "three" in feeling:
        print (f"""
        That's understandable, this is a whole new experience... What if I tell you
        that you stand really good chances of completing this mission.
        
        Will that give you more hope?\n
        < Please type Y/N >\n""")
        
        ###User's input for insecure expectations
        motiviation_answer= input(prompt= "")
        
        ###Analizing user's answer - nested conditional
        ###This condition won't affect the path that the user will follow, it will only try to empathize
        if motiviation_answer == 'yes' or motiviation_answer == "Y" or "y" in motiviation_answer :
            print("""\n\nHooray!!! That's such a good attitude!!!\n""")
        else :
            print("""\n\nWell, hopefully you'll be more enthusiastic as you spend more time on space.\n""")
    
    ###If the user inputs sth different than what was asked, they will still continue with the game...
    ###...but the message will be different
    else :
        print (f"""\nI'm sorry, I didn't get how you feel""")
    ###Sending user to the next stage after a message according to its feelings was displayed
    #out_of_vialactea()
    print("End of second stage")
#################################################################################
################################## OUT OF VIA LACTEA ############################
traveling_space()
```
